# Checking M31 17B Data

Up to now (10/09/2017), 20 hours observation has been executed in 17B (total 25 hours).

This note is used to check weather our 15 hours Band 2 observation towards Region-1 (galactic center) is deep enough.

# 450 um Band

Plot 17A Region-1 450 um image. (3 hours Band 2 + 5 hours Band 3 observing time)

Using DS9, we could measure a rms noise of ~22 mJy/beam from the image center. (considering a angule resolution of 8")

```python
import matplotlib
from astropy import units as u
matplotlib.use('Agg')
import aplpy
%matplotlib inline
```

```python
f_17a = aplpy.FITSFigure('/Users/shjiao/desktop/region12/17b450region1/m17A_450.conv.fits', figsize=(7,7))
f_17a.show_grayscale()
f_17a.add_colorbar()
f_17a.colorbar.show()
f_17a.recenter(10.6403042,41.35090276,width=0.27,height=0.27)
f_17a.colorbar.set_location('right')
f_17a.colorbar.set_axis_label_text('Flux (mJy/beam)')
f_17a.add_label(0.87,0.85, 'M17A SCUBA2 \n450 $\mu m$ ', color='red', relative=True, size='large',layer='source')
f_17a.show_colorscale(vmin=-30,vmax=60,cmap='gist_yarg',stretch='linear')
f_17a.add_beam(major=8.*u.arcsecond,minor=8.*u.arcsecond,angle=0.0)
f_17a.add_label(0.1, 0.08, 'Beam: 8\"', relative=True, size='large', layer='title',color='red')
f_17a.tick_labels.set_font(size='xx-small')
f_17a.axis_labels.set_font(size='x-small')
```

Plot the extropolated 450 um image, which comes from the SED fitting based on Herschel dataset.

The red lines trace 33 mJy/beam, the blue lines trace 22 mJy/beam.

Now, we only can get S/N >3 within the red contour. But we plan to use 15 hours Band 2 observing time to get a rms noise of ~7.6 mJy/beam, and in this way, we could get S/N >3 within the blue contour.

The existing 450 um Region-1 image doesn't reach the sensitivity what we want, and I think we should use the remaining 5 hours towards Region-1. 

```python
f_extra = aplpy.FITSFigure('/Users/shjiao/desktop/m31/sedto450/flux450.conv.fits', figsize=(7,7))
f_extra.show_grayscale()
f_extra.add_colorbar()
f_extra.colorbar.show()
f_extra.recenter(10.6403042,41.35090276,width=0.27,height=0.27)
f_extra.colorbar.set_location('right')
f_extra.colorbar.set_axis_label_text('Flux (mJy/beam)')
f_extra.add_label(0.8,0.9, ' extrapolated \n450 $\mu m$', color='black', relative=True, size='large',layer='source')
f_extra.show_colorscale(cmap='gist_yarg')
f_extra.show_contour('/Users/shjiao/desktop/m31/sedto450/flux450.conv.fits',levels=[22.,33.],width='2')
f_extra.tick_labels.set_font(size='xx-small')
f_extra.axis_labels.set_font(size='x-small')
```
